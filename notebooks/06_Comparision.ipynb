```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
svm = pd.read_csv(
    "/content/drive/MyDrive/FP_AI_2026/compare/svm_results.csv"
)

cnn = pd.read_csv(
    "/content/drive/MyDrive/FP_AI_2026/compare/cnn_results.csv"
)

graph = pd.read_csv(
    "/content/drive/MyDrive/FP_AI_2026/compare/graphcodebert_results.csv"
)
```

```python
comparison_df = pd.concat(
    [
        svm,
        cnn,
        graph
    ],
    ignore_index=True
)

comparison_df
```

```python
comparison_df.to_csv(
    "/content/drive/MyDrive/FP_AI_2026/compare/model_comparison.csv",
    index=False
)

print(
    "Comparison table saved!"
)
```

```python
comparison_df.sort_values(
    by="Accuracy",
    ascending=False
)
```

```python
plt.figure(
    figsize=(8,5)
)

sns.barplot(
    data=comparison_df,
    x="Model",
    y="Accuracy"
)

plt.title(
    "Accuracy Comparison"
)

plt.show()
```

```python
plt.figure(
    figsize=(8,5)
)

sns.barplot(
    data=comparison_df,
    x="Model",
    y="F1"
)

plt.title(
    "F1 Score Comparison"
)

plt.show()
```

```python
plt.figure(
    figsize=(8,5)
)

sns.barplot(
    data=comparison_df,
    x="Model",
    y="AUC"
)

plt.title(
    "AUC Comparison"
)

plt.show()
```

```python
plt.figure(
    figsize=(8,5)
)

sns.barplot(
    data=comparison_df,
    x="Model",
    y="Training_Time"
)

plt.title(
    "Training Time Comparison"
)

plt.ylabel(
    "Seconds"
)

plt.show()
```

```python
plt.figure(
    figsize=(8,5)
)

sns.barplot(
    data=comparison_df,
    x="Model",
    y="Inference_Time"
)

plt.title(
    "Inference Time Comparison"
)

plt.ylabel(
    "Seconds"
)

plt.show()
```

```python
plt.figure(
    figsize=(8,5)
)

sns.barplot(
    data=comparison_df,
    x="Model",
    y="Parameters"
)

plt.title(
    "Model Parameters Comparison"
)

plt.ylabel(
    "Number of Parameters"
)

plt.show()
```

```python
metrics_df = comparison_df.set_index(
    "Model"
)[
    [
        "Accuracy",
        "Precision",
        "Recall",
        "F1",
        "AUC"
    ]
]

plt.figure(
    figsize=(10,5)
)

sns.heatmap(
    metrics_df,
    annot=True,
    cmap="Blues",
    fmt=".4f"
)

plt.title(
    "Performance Comparison Heatmap"
)

plt.show()
```

```python
best_model = comparison_df.loc[
    comparison_df["Accuracy"].idxmax()
]

print(
    "Best Model:"
)

display(best_model)
```
