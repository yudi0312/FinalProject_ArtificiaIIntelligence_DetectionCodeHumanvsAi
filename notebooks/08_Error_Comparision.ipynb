```python
import pandas as pd
```

```python
svm = pd.read_csv(
    "/content/drive/MyDrive/FP_AI_2026/compare/svm_error_summary.csv"
)

cnn = pd.read_csv(
    "/content/drive/MyDrive/FP_AI_2026/compare/cnn_error_summary.csv"
)

graph = pd.read_csv(
    "/content/drive/MyDrive/FP_AI_2026/compare/graphcodebert_error_summary.csv"
)

comparison = pd.concat(
    [
        svm,
        cnn,
        graph
    ],
    ignore_index=True
)

comparison
```

```python
comparison.to_csv(
    "/content/drive/MyDrive/FP_AI_2026/compare/error_analysis_comparison.csv",
    index=False
)

comparison
```

```python
import matplotlib.pyplot as plt

comparison.set_index(
    "Model"
)[
    [
        "False_Positive",
        "False_Negative"
    ]
].plot(
    kind="bar",
    figsize=(7,4)
)

plt.ylabel(
    "Jumlah Error"
)

plt.title(
    "False Positive dan False Negative"
)

plt.grid(
    axis="y"
)

plt.show()
```
